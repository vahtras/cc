```python
# %load demo.py
import pandas as pd

df = pd.read_csv(
    'https://docs.google.com/spreadsheets/d/'
    '1sg97nDTQ0VN5_2NtDLR-BKfCIhv4jHXoUT_y4ApOdJ0/'
    'export?format=csv'
).fillna('')
```

```python
df
```

```python
dfstat = df.groupby(['Col1', 'Col4'])['col12'].sum().reset_index()
dfstat
```

```python
country = df[['Col4', 'col5']].drop_duplicates().set_index('Col4')
country
```

```python
dfstat['country'] = dfstat['Col4'].apply(lambda cache: country.loc[cache])
dfstat = dfstat[['Col1', 'Col4', 'country', 'col12']]
dfstat
```
